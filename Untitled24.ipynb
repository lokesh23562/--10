```python
# ==========================================
# 1. IMPORT LIBRARIES
# ==========================================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

# ==========================================
# 2. LOAD DATA
# ==========================================
df = pd.read_csv("/content/sample_data/customer_churn (4).csv")

print("Dataset Preview:")
print(df.head())

# ==========================================
# 3. DATA PREPROCESSING
# ==========================================
# Handle categorical data if any
df = pd.get_dummies(df, drop_first=True)

# Separate features
X = df.drop("Churn", axis=1)

# Scale data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# ==========================================
# 4. ELBOW METHOD (K-MEANS)
# ==========================================
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=42)
    kmeans.fit(X_scaled)
    wcss.append(kmeans.inertia_)

plt.figure()
plt.plot(range(1, 11), wcss, marker='o')
plt.title("Elbow Method")
plt.xlabel("Number of Clusters")
plt.ylabel("WCSS")
plt.show()

# ==========================================
# 5. APPLY K-MEANS
# ==========================================
kmeans = KMeans(n_clusters=3, random_state=42)
df["Cluster"] = kmeans.fit_predict(X_scaled)

# ==========================================
# 6. HIERARCHICAL CLUSTERING
# ==========================================
hc = AgglomerativeClustering(n_clusters=3)
df["HC_Cluster"] = hc.fit_predict(X_scaled)

# ==========================================
# 7. DBSCAN CLUSTERING
# ==========================================
db = DBSCAN(eps=0.5, min_samples=5)
df["DBSCAN_Cluster"] = db.fit_predict(X_scaled)

# ==========================================
# 8. VISUALIZATION
# ==========================================
plt.figure()
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=df["Cluster"], palette="viridis")
plt.title("Customer Segments (K-Means)")
plt.show()

# ==========================================
# 9. SEGMENT ANALYSIS
# ==========================================
print("\nCluster Summary:")
print(df.groupby("Cluster").mean())

# ==========================================
# 10. MODEL BUILDING PER SEGMENT
# ==========================================
models = {}
results = []

for cluster in df["Cluster"].unique():
    print(f"\nProcessing Cluster {cluster}...")

    segment = df[df["Cluster"] == cluster]

    if len(segment) < 10:
        print("Skipping small cluster")
        continue

    X_seg = segment.drop(["Cluster", "HC_Cluster", "DBSCAN_Cluster", "Churn"], axis=1)
    y_seg = segment["Churn"]

    X_train, X_test, y_train, y_test = train_test_split(
        X_seg, y_seg, test_size=0.2, random_state=42
    )

    # ==========================================
    # 11. HYPERPARAMETER TUNING
    # ==========================================
    param_grid = {
        'n_estimators': [50, 100],
        'max_depth': [5, 10, None]
    }

    grid = GridSearchCV(RandomForestClassifier(), param_grid, cv=3)
    grid.fit(X_train, y_train)

    best_model = grid.best_estimator_
    models[cluster] = best_model

    # ==========================================
    # 12. PREDICTIONS
    # ==========================================
    y_pred = best_model.predict(X_test)

    # ==========================================
    # 13. EVALUATION METRICS
    # ==========================================
    acc = accuracy_score(y_test, y_pred)
    prec = precision_score(y_test, y_pred, zero_division=0)
    rec = recall_score(y_test, y_pred, zero_division=0)
    f1 = f1_score(y_test, y_pred, zero_division=0)

    try:
        roc = roc_auc_score(y_test, best_model.predict_proba(X_test)[:,1])
    except:
        roc = 0

    results.append({
        "Cluster": cluster,
        "Accuracy": acc,
        "Precision": prec,
        "Recall": rec,
        "F1": f1,
        "ROC-AUC": roc
    })

    # ==========================================
    # 14. FEATURE IMPORTANCE
    # ==========================================
    importance = pd.Series(best_model.feature_importances_, index=X_seg.columns)
    print("\nFeature Importance:")
    print(importance.sort_values(ascending=False))

# ==========================================
# 15. RESULTS SUMMARY
# ==========================================
results_df = pd.DataFrame(results)

print("\nFinal Model Performance:")
print(results_df)

# Save results
results_df.to_csv("model_evaluation_results.csv", index=False)

# ==========================================
# 16. SAVE SEGMENTED DATA
# ==========================================
df.to_csv("segmentation_data.csv", index=False)

# ==========================================
# 17. BUSINESS INSIGHTS
# ==========================================
print("\nBUSINESS INSIGHTS:")

for cluster in df["Cluster"].unique():
    print(f"\nCluster {cluster} Strategy:")

    if cluster == 0:
        print("Premium Customers → Offer loyalty programs & exclusive deals")
    elif cluster == 1:
        print("Budget Customers → Provide discounts and coupons")
    else:
        print("Regular Customers → Improve engagement and retention offers")

print("\nProject Completed Successfully!")
```
